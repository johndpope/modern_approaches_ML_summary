```python
import numpy as np

p = np.random.randint( 3, 10 )
q = np.random.randint( 3, 10 )
print( "PQ", p, q)
    
r = min( p, q )

# select randomly a column from the initial matrix (to be removed)
k = np.random.randint( q )
k = q-1


# construct a random matrix of size m, n
A = np.random.randn( p, q )
a = -A[ :, k ].reshape( p,1 )
b = np.zeros( (q,1) ); b[ k ] = 1

Am = A[ :, np.delete( np.arange(0, q), k ) ]



# retrieve a first SVD representation
U, s, Vt = np.linalg.svd( A, full_matrices=False )
S = np.diag( s )

print( "FI", U.shape, s.shape, Vt.shape )


# construct the K matrix and its SVD representation
nv = np.dot( Vt, b )
Q = (b - np.dot( np.transpose(Vt), nv )) / np.sqrt( 1-np.dot( np.transpose(nv), nv ) )
Q = np.nan_to_num( Q )

print( "Q", Q )


Sx = np.diag( np.append( s, 0 ) )        

K = np.dot(

    Sx,
    (
        +np.identity( r+1 )
        -np.dot(
            np.append( np.dot( S, nv ), np.zeros( (1,1) ), axis= 0 ),
            np.transpose( np.append( nv, np.sqrt( 1-np.dot( np.transpose(nv), nv ) ), axis=0 ) )
        )
    )
); K = np.nan_to_num( K )


print( "\n\n\nK", K )
Up, sp, Vpt = np.linalg.svd( K, full_matrices=False )


iU = Up[ :-1, :-1 ]
s = sp[ :-1 ]
iVt = Vpt[ :-1, : ]

Vt = np.transpose( np.append( np.transpose( Vt ), Q, axis=1 ) )
Vt = Vt[ :, np.delete( np.arange(0, q, 1), k ) ]

U = np.dot( U, iU )
Vt = np.dot( iVt, Vt )


print( "N1", np.linalg.norm( np.dot( U.T, U ) - np.eye( U.shape[ 1 ]) ) )    
print( "N2", np.linalg.norm( np.dot( Vt.T, Vt ) - np.eye( Vt.shape[ 1 ]) ) ) 
print( "N3", np.linalg.norm( Am - np.dot( U, np.dot( np.diag(s), Vt ) ) ) )
print( "N4", np.linalg.norm( np.dot( Am, np.transpose( Vt ) ) - np.dot( U, np.diag(s) ) ) )

print( "N5", np.linalg.norm( np.dot( np.transpose( U ), Am ) - np.dot( np.diag(s), Vt ) )/np.linalg.norm( Am ) )


uf, sf, vf = np.linalg.svd( Am, full_matrices=False)
print( "K.shape", K.shape )
print("F", uf.shape, sf.shape, vf.shape )
print("F", U.shape, s.shape, Vt.shape )

print( "N4", np.linalg.norm( np.dot( Am, np.transpose( vf ) ) - np.dot( uf, np.diag(sf) ) ) )





Ai = np.dot( U, np.dot( np.diag( s ), Vt ) )

print( "K", k, "\nA", A, "\nAi", Ai,  )
print( "Matrix: ", np.allclose( Ai, Am ) )


sx = s.copy()
nzindx = sx > 0; sx[ nzindx ] = 1.0/sx[ nzindx ]
print( s, sx )

Ainv = np.dot( Vt.T, np.dot( np.diag( sx ), U.T ) )



print( "Final ", np.dot( Ainv,  Am ) )
print( "Errro: ", np.linalg.norm(  np.dot( Am, np.dot( Ainv,  Am ) ) - Am ) )
print( "AError", np.allclose(  np.dot( Am, np.dot( Ainv,  Am ) ),  Am ) )

def svdsolve(Ainp):
    u, s, v = np.linalg.svd(Ainp, full_matrices=False)
    Ainvo = np.dot (v.T, np.dot(np.diag(s**-1),u.T))
    return(Ainvo)

print( "INvC", np.dot( svdsolve(Am), Am ) )
print( "Error: ", np.linalg.norm(  np.dot( Am, np.dot( svdsolve(Am), Am ) ) - Am ) )
print( "PQ", p, q)
```
